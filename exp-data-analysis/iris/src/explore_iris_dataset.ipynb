```julia
client = Mongoc.Client()
```

```julia
Mongoc.ping(client)
```

```julia
database = client["my-database"]
```

```julia
collection = database["my-collection"]
```

```julia
using JSON
```

```julia
datadict = JSON.parse(JSON.json(test_data))
```

```julia
datadict["id"] = "iris_test_data"
```

```julia
document = Mongoc.BSON(datadict)
```

```julia
result = push!(collection, document)
```

```julia
data_from_mongo = Mongoc.find_one(collection, document)
```
